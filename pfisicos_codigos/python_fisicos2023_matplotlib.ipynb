```python
#Marcadores - Exemplo
# Vamos plotar uma função periódica simulando um osciloscópio
def tophat(rx):
    """função: y = 1 for x < 0.5, y=0 for x >= 0.5 """
    ry = np.ones(rn)
    ry[rx>=0.5]=0
    return ry

rn = 100
rx = np.linspace(0, 1, rn)

# Repetir a função nrep vezes
nrep = 4
x = np.linspace(0, nrep, nrep*rn)
y = np.tile(tophat(rx), nrep)

#Desenhar a figura a modificar os marcadores
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x,y, 'k', lw=2)
ax.set_ylim(-0.1,1.1)
ax.set_xlim(x[0]-0.5, x[-1]+0.5)
# Marcadores
ax.minorticks_on()
ax.tick_params(which='major', length=10, width=2, labelsize=13, direction='inout',color='r')
ax.tick_params(which='minor', length=5, width=2, labelsize=13, direction='in',color='b')
ax.grid(which='both',ls='--')
```

**Exemplo de Aplicação: Dados do Higgs**

```python
data_higgs = np.loadtxt('higgs_data.csv',delimiter=',')
data_higgs[:5]
```

```python
data_higgs.shape
```

```python
lbkg = [24415.1,-356.488,1.80183,-0.00307196]
lsig = [200,126.218,2.3875]
```

```python
mass = data_higgs[:,0]
events = data_higgs[:,1]
```

```python
def bkg(m,par_bkg):
    d,c,b,a = par_bkg
    return a*m**3 + b*m**2 + c*m +d    
```

```python
def signal(m,par_sig):
    A,mu,sigma = par_sig
    return A*np.exp(-0.5*((m-mu)**2/sigma**2))
```

```python
m = np.linspace(mass.min(),mass.max(),200)
ftotal = bkg(m,lbkg) + signal(m,lsig)
```

```python
fig,ax = plt.subplots()
ax.plot(mass,events,'ok',label='Data')
ax.plot(m,bkg(m,lbkg),'r--',label='Bkg')
ax.plot(m,ftotal,'r-',lw=2,label='Sig + Bkg')
ax.set_xlabel(r'$m_{\gamma\gamma}$',fontsize=14,labelpad=14)
ax.set_ylabel('Events',fontsize=14,labelpad=15)
ax.tick_params(labelsize=13)
ax.legend(frameon=False)
ax.text(105,2000,r'$H \rightarrow \gamma\gamma$',fontsize=14)
```

**Gráficos 3D**

```python
x = np.linspace(-2, 2, 400)
y = x.copy()
X, Y = np.meshgrid(x, y)
Z = np.exp(-(X**2 + Y**2))
fig, ax = plt.subplots(figsize=(8,7),nrows=2, ncols=2, subplot_kw={'projection': '3d'})
ax[0,0].plot_wireframe(X, Y, Z, rstride=40, cstride=40)
ax[0,1].plot_surface(X, Y, Z, rstride=40, cstride=40, color='r')
ax[1,0].plot_surface(X, Y, Z, rstride=12, cstride=12, color='m')
ax[1,1].plot_surface(X, Y, Z, rstride=20, cstride=20, cmap='hot')
for axes in ax.flatten():
    axes.set_xticks([-2, -1, 0, 1, 2])
    axes.set_yticks([-2, -1, 0, 1, 2])
    axes.set_zticks([0, 0.5, 1])
fig.tight_layout()
```
